```python
import numpy as np
import matplotlib.pyplot as plt
from model import Gentrification
from agents import House

gentri = Gentrification(density=0.6, width= 20, height=20, immigrant_start=100, immigrant_count=160, income_variance=1, preference=0.5)

# Run the model for 300 timesteps
for _ in range(300):
    gentri.step()

locational_quality_grid = np.zeros((gentri.grid.width, gentri.grid.height))
for cell in gentri.grid.coord_iter():
    cell_content, (x, y) = cell
    # Find the House agent in the cell content
    house = next((agent for agent in cell_content if isinstance(agent, House)), None)
    if house:
        locational_quality_grid[x][y] = house.locational_quality

# Create a heatmap to visualize the locational_quality distribution
plt.imshow(locational_quality_grid, cmap='hot', interpolation='nearest')
plt.colorbar(label='Locational Quality')
plt.title('Distribution of Locational Quality')
plt.xlabel('Grid X')
plt.ylabel('Grid Y')
plt.show()
```
